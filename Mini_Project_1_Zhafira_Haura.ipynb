##Lambda Function

```python
df
```

```python
#ingin buat kolom baru namanya "Category"
#isinya berupa category hp berdasarkan harga
#jika harganya > $500, maka categorynya MAHAL
#jika harganya < $500, maka category MURAH

#siapkan fungsi penetapan category hp berdasarkan harganya
def get_category(price):
    category = ''
    if price <= 500:
        category = 'MURAH'
    else:
        category = 'MAHAL'
    return category
```

```python
get_category(1000)
```

```python
get_category(200)
```

```python
#lakukan iterasi untuk setiap record di DataFrame
#cek dan tetapkan category hp menggunakan fungsi get_category

for i in range(len(df)):
    print(get_category(df.loc[i, 'Price ($)']))
```

```python
for i in range(len(df)):
    df.loc[i, 'category'] = get_category(df.loc[i, 'Price ($)'])
```

```python
df
```

```python
#manfaatkan lambda function
df['category_lambda'] = df['Price ($)'].apply(lambda price: 'MURAH' if price <= 500 else 'MAHAL')
```
